```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
```

```python
x = np.linspace(0,5,11)
y = x ** 2
```

```python
x
```

```python
y
```

```python
# Functional Method
plt.plot(x,y)
```

```python
plt.plot(x,y,'r-') #marker colors
```

```python
plt.plot(x,y)
plt.xlabel("X Label")
plt.ylabel('Y Label')
plt.title('Title')
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r')

plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

```python
#object oriented method
fig = plt.figure()
axes = fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(x,y)
axes.set_xlabel('X label')
axes.set_ylabel('Y label')
axes.set_title('Set Title')
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])

axes1.plot(x,y)
axes1.set_title('Larger Plot')
axes2.plot(y,x)
axes2.set_title('Smaller Plot')
```

```python
fig,axes = plt.subplots(nrows=3,ncols=3)
plt.tight_layout()
#axes.plot(x,y)
```

```python
fig,axes = plt.subplots(nrows=1,ncols=2)
#so axes contains an array of matplotlib objects
#so we can iterate over it
for current in axes:
    current.plot(x,y)
```

```python
fig,axes = plt.subplots(nrows=1,ncols=2)
axes[0].plot(x,y)
axes[0].set_title('First Plot')
axes[0].set_xlabel('X label')
axes[0].set_ylabel('Y label')

axes[1].plot(y,x)
axes[1].set_title('Second Plot')
axes[1].set_xlabel('X label')
axes[1].set_ylabel('Y label')
plt.tight_layout() #takes care of overlapping
```

### Figure size and DPI

```python
fig = plt.figure(figsize=(8,2)) # width , height
ax= fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
fig,axes = plt.subplots(nrows=2,ncols=1,figsize=(8,2))
axes[0].plot(x,y)
axes[1].plot(y,x)
plt.tight_layout()
```

### Legends

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,x ** 2,label='X Squared')
ax.plot(x,x ** 3,label='X Cubed')

#adding legends 
ax.legend(loc=0) #adding location is not necessary
```

### Plot Appearance

```python
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])
axes.plot(x,y,color='orange',linewidth=3,alpha = 0.5,linestyle='--',marker='o',markersize=10,
         markerfacecolor='yellow',markeredgewidth=3,markeredgecolor='green') # RGB hex codes are also possible
# transparency is set by changing alpha
```

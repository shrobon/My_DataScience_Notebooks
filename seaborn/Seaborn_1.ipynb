### Distribution Plots

```python
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import pandas as pd
import numpy as np

```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head(6)
```

```python
sns.distplot(tips['total_bill'],kde=False,bins=15)
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips) # by default it is scatter
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='hex')
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='reg')
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='kde')
```

```python
sns.pairplot(tips) #does it for every column, otherwise same as jointplot
```

```python
sns.pairplot(tips,hue='sex')
```

```python
sns.pairplot(tips,hue='sex',palette='coolwarm')
```

```python
sns.rugplot(tips['total_bill'])
```

```python
sns.distplot(tips['total_bill'],kde=False)
```

```python
# KDE - stands for kernal density estimate plots
# has something to do with rugplots 
# See Wikipedia
```

```python
sns.kdeplot(tips['total_bill'])
```
